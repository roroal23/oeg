## BLEU score

```python
from nltk.translate import bleu_score
```

```python
def read_lines(f):
    line1 = f.readline().strip("\n")
    if not line1:
        return None
    line2 = f.readline().strip("\n")
    line3 = f.readline().strip("\n")
    return [line1, line2, line3]
```

```python
def split_and_pack(lines: list[str]):
    reference1 = lines[0].split()
    reference2 = lines[1].split()
    hypothesis = lines[2].split()
    return [[reference1, reference2], hypothesis]
```

```python
def bleu_step_by_step(references, hypothesis, weights, smoothing_function = None) -> float:

    hyp_len = len(hypothesis)
    closest_ref_length = bleu_score.closest_ref_length(references, hyp_len)
    print(f"Closest reference length: {closest_ref_length}")
    brevity_penalty = bleu_score.brevity_penalty(closest_ref_length, hyp_len)
    print(f"Brevity penalty: {brevity_penalty}")
    corpus_bleu = bleu_score.corpus_bleu([references], [hypothesis], weights)
    print(f"Corpus bleu: {corpus_bleu}")
    sentence_bleu = bleu_score.sentence_bleu(references, hypothesis, weigths)
    print(f"Sentence bleu: {sentence_bleu}")
    return sentence_bleu
```

```python
#Prueba inicial con BLEU
reference1 = "El perro come pienso de pavo".split()
reference2 = "El perro consume pienso de pavo".split()
hypothesis = "El perro cena pienso de pavo".split()


references = [reference1, reference2]
weigths = (0.5, 0.5, 0, 0)

bleu_step_by_step(references, hypothesis, weigths)
```

```python
#Como las oraciones son sencillas no hay 3-gramas ni 4-gramas. Sus pesos se inicializan a 0 para que el score tenga un valor significativo (seria ~0 en otro caso)
weigths = (0.5, 0.5, 0, 0)

with open("mini_dataset.txt", 'r') as f:
    while True:
        print("-------------------------------------------")
        lines = read_lines(f)
        if not lines:
            break
        print(f"References: {lines[0]}, {lines[1]}")
        print(f"Hypothesis: {lines[2]}")
        references, hypothesis = split_and_pack(lines)
        bleu_step_by_step(references, hypothesis, weigths)
```

## ROUGE

### rouge_score

```python
from rouge_score import rouge_scorer
```

```python
scorer = rouge_scorer.RougeScorer(['rouge1','rouge2','rougeL'], use_stemmer=True)
```

```python
def print_rouge_results(results):
    for metric, score in results.items():
        print(f"{metric}:")
        print(f"    Precision: {score.precision}")
        print(f"    Recall: {score.recall}")
        print(f"    F1: {score.fmeasure}")
```

```python
def rouge_step_by_step(reference, hypothesis):
    print(f"Reference: {reference}")
    print(f"Hypothesis: {hypothesis}")
    scorer_no_stemmer = rouge_scorer.RougeScorer(['rouge1','rouge2','rougeL'], use_stemmer=False)
    results_no_stemmer = scorer_no_stemmer.score(reference1, hypothesis)
    scorer = rouge_scorer.RougeScorer(['rouge1','rouge2','rougeL'], use_stemmer=True)
    results = scorer.score(reference1, hypothesis)
    print("---No Stemmer---")
    print_rouge_results(results_no_stemmer)
    print("---With Stemmer---")
    print_rouge_results(results)


```

```python
reference1 = "El perro come pienso de pavo"
hypothesis = "El perro cena pienso de pavo"
scorer = rouge_scorer.RougeScorer(['rouge1','rouge2','rougeL'], use_stemmer=True)
print_rouge_results(scorer.score(reference1, hypothesis))
```

#### Uso del stemmer

```python
reference1 = "Playing cooking dancing participating walking"
reference2 = "Played cooked danced participated walked"
hypothesis = "Play cook dance participate walk"
rouge_step_by_step(reference1, hypothesis)
rouge_step_by_step(reference2, hypothesis)
```
